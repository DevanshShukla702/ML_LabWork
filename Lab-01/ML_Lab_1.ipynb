```python
import numpy as np
data_list = [10, 20, 30, 40, 50]
arr = np.array(data_list)

print("Array:", arr)
print("Shape:", arr.shape)
print("Data Type:", arr.dtype)
```

```python
zeros_matrix = np.zeros((3, 4))
ones_matrix = np.ones((2, 5))

print("Zeros Matrix:\n", zeros_matrix)
print("Shape:", zeros_matrix.shape)
print("\nOnes Matrix:\n", ones_matrix)
print("Shape:", ones_matrix.shape)
```

```python

original = np.arange(1, 13)
reshaped = original.reshape(3, 4)

print("Original Array:", original)
print("Original Shape:", original.shape)
print("\nReshaped Matrix:\n", reshaped)
print("Reshaped Shape:", reshaped.shape)
```

```python
rand_arr = np.random.randint(1, 100, 10)

print("Array:", rand_arr)
print("Min:", rand_arr.min())
print("Max:", rand_arr.max())
print("Mean:", rand_arr.mean())
print("Sum:", rand_arr.sum())
```

```python
import pandas as pd

df = pd.read_csv('/content/sample_data/california_housing_train.csv')
summary = df.describe()

print("--- Statistical Summary of the Dataset ---")
print(summary)
target_col = df.select_dtypes(include=['number']).columns[0]

print(f"\n--- Interpretation for Column: {target_col} ---")
print(f"Mean (Average): {summary.loc['mean', target_col]:.2f}")
print(f"Standard Deviation (Spread): {summary.loc['std', target_col]:.2f}")
print(f"Minimum Value: {summary.loc['min', target_col]}")
print(f"Maximum Value: {summary.loc['max', target_col]}")
```
