```python
import pandas as pd

# Load dataset
df = pd.read_csv('/StudentsPerformance.csv')

# Display first few records
print("First 5 records:\n", df.head())

# Identify numerical and categorical columns
print("\nColumn Data Types:\n", df.dtypes)

# Selecting a numerical feature for analysis: 'math score'
target = df['math score']

# Measures of Central Tendency
print("\n--- Measures of Central Tendency ---")
print(f"Mean: {target.mean():.2f}")
print(f"Median: {target.median()}")
print(f"Mode: {target.mode()[0]}")

# Measures of Dispersion
print("\n--- Measures of Dispersion ---")
print(f"Minimum: {target.min()}")
print(f"Maximum: {target.max()}")
print(f"Sum: {target.sum()}")
print(f"Variance: {target.var():.2f}")
print(f"Standard Deviation: {target.std():.2f}")
```

```python
# Computing Quartiles
q1 = df['math score'].quantile(0.25)
q2 = df['math score'].quantile(0.50)  # Median
q3 = df['math score'].quantile(0.75)

print(f"First Quartile (Q1): {q1}")
print(f"Second Quartile (Q2/Median): {q2}")
print(f"Third Quartile (Q3): {q3}")

# Interpretation:
# Q1: 25% of students scored below {q1}.
# Q3: 75% of students scored below {q3}, meaning the top 25% scored above this.
```

```python
# Correlation and Covariance between Math and Reading scores
correlation = df['math score'].corr(df['reading score'])
covariance = df['math score'].cov(df['reading score'])

print(f"Correlation between Math and Reading: {correlation:.4f}")
print(f"Covariance between Math and Reading: {covariance:.4f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot Histogram
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
sns.histplot(df['math score'], kde=True, color='blue')
plt.title('Math Score Distribution (Histogram)')

# Plot Boxplot
plt.subplot(1, 2, 2)
sns.boxplot(x=df['math score'], color='green')
plt.title('Math Score Quartiles & Outliers (Boxplot)')

plt.tight_layout()
plt.show()
```
